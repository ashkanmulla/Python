```python
#Changing the type of a column
df.mpg = df.mpg.astype(float)
#See the changes
df.info()
```

```python
#Filling the empty cells with mean
df.qsec = df.qsec.fillna(df.qsec.mean())
df
```

```python
#Drop an unwanted column
df = df.drop(columns=["am"])
```

```python
#Sorting
df["hp"] = 10
```

```python
#Apply Function
f = lambda x : x+10
df["cyl"] = df["cyl"].apply(f)
```

```python
#Sorting cyl column in ascending order
newvar = df.sort_values(by= "qsec")
```

```python
newvar.head()
```

```python
#Sorting cyl column in decending order
newvar = df.sort_values(by= "qsec",ascending= False)
newvar.head()
```
